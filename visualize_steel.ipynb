```python
import numpy as np
import matplotlib.pyplot as plt
from os import path

steel_data_dir = path.abspath(path.join("data", "steel_defect", "processed"))
steel_images = path.join(steel_data_dir, "images")
steel_labels = path.join(steel_data_dir, "labels")
```

```python
IMG_OF_INTEREST = "000f6bf48"
```

```python
fname = IMG_OF_INTEREST + ".npy"
img_array = np.load(path.join(steel_images, fname))
label_array = np.load(path.join(steel_labels,fname))
```

The label has 4 channels. At most 2 of them are ever used, but we can only display
three: drop a channel that has all zeroes.

```python
channel_sums = label_array.reshape(-1,4).sum(axis = 0)
print(channel_sums)
drop_channel = list(channel_sums).index(0)

print("Dropping channel", drop_channel)
label_reduced = label_array[:,:,[i != drop_channel for i in range(4)] ]
```

```python
def plot_img(img: np.ndarray):
    plt.figure()
    plt.imshow(img)
    plt.show()
    plt.close()
```

```python
plot_img(img_array)
```

```python
plt.imshow(label_reduced)
```
